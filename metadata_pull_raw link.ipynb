```python
import requests
```

```python
files = [
    {
        'name':'dim_calender',
        'url':'https://raw.githubusercontent.com/pradeep-1004/git_API_test_dataset/main/AdventureWorks_Calendar.csv' 
    },
    {
        'name':'dim_products',
        'url':'https://raw.githubusercontent.com/pradeep-1004/git_API_test_dataset/main/AdventureWorks_Products.csv'
    },
    {
        'name':'dim_product_category',
        'url':'https://raw.githubusercontent.com/pradeep-1004/git_API_test_dataset/main/AdventureWorks_Product_Categories.csv'
    }
]
```

```python
bronze_base='/Volumes/parent_cata/git_pull/bronze_metadata_raw'
```

```python
#loop
for item in files:
    print(f'processing:{item['name']}')

    response = requests.get(item['url'])
    data = response.text.splitlines()



    bronze_target_path = f'{bronze_base}/{item['name']}'

    df = spark.createDataFrame(data)
    
    df.write.format('delta')\
        .option('overwriteSchema', 'true')\
        .mode('overwrite')\
        .save(bronze_target_path)

    print(f'writing to {bronze_target_path}')
    
```
